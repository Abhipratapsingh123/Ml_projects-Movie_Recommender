```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
import ast
```

```python
# Loading the dataset
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Merging both the datasets
movies = movies.merge(credits,on ='title')
```

```python
movies.head(2)
```

```python
# columns to keep
# Genre
# ID
# keywords
# title
# overview
# cast
# crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# checking for null values
movies.isnull().sum()
```

```python
# dropping those 3 null values
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
# checking duplicate values
movies.duplicated().sum()
```

```python
# function to fetch out the 'name' key from the dictionary in the genre column

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# applying to the genre column
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

#### **Applying same preprocessing steps to keyword column**

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Converting overview column from string to list
movies['overview']= movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# Now removing space from every word in order to create useful tags

movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# creating new column tag in the movies dataset
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
# now creating new dataframe only with required columns
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# converting tags column from list to string
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
# converting to lower case

new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
# performing stemming on tags column
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# stemmed new_df
new_df.head()
```

```python
# vectorizing the text to create bag of words

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# checking top 5000 words
cv.get_feature_names_out()
```

#### **Now calculating cosine distances b/w each vector, more the distance less the similarity b/w them**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

#### **Now making a recommended function**

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

### **Recommnending Movies based on given movie**

```python
recommend("Avatar")
```
